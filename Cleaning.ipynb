# Clean

```python
Railway_Clean = Railway.copy()
```

## Inconsistent Data

### Define:
Change Date of Purchase and Date of Journey Columns Format

### Code:

```python
Railway_Clean['Date of Purchase'] = pd.to_datetime(Railway['Date of Purchase'])
Railway_Clean['Date of Journey'] = pd.to_datetime(Railway['Date of Journey'])
```

### Test:

```python
Railway_Clean.info()
```

```python
Railway_Clean.sample(10)
```

## Missing Data

### Define:
Complete the Data in Railcard Column

### Code:

```python
Railway_Clean['Railcard'].fillna('Not Subscribed', inplace = True)
```

### Test:

```python
Railway_Clean.info()
```

```python
Railway_Clean.sample(10)
```

### Define:
Complete the Data in Reason for Delay Column

### Code:

```python
Railway_Clean['Reason for Delay'].fillna('No Delay', inplace = True)
```

Test:

```python
Railway_Clean.info()
```

```python
Railway_Clean.sample(10)
```

## Repeated Data

### Define:


```python
Railway_Clean['Reason for Delay'] = Railway_Clean['Reason for Delay'].str.replace('Weather Conditions', 'Weather')
Railway_Clean['Reason for Delay'] = Railway_Clean['Reason for Delay'].str.replace('Signal failure', 'Signal Failure')
Railway_Clean['Reason for Delay'] = Railway_Clean['Reason for Delay'].str.replace('Staffing', 'Staff Shortage')
```
