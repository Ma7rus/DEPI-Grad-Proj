# Analysis Questions

## Q1/ what are the numbers of each railcard type and their sales?

### Code:

```python
railcard_counts = Railway['Railcard'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(y=railcard_counts.index, x=railcard_counts.values, orient='h')
plt.ylabel('Railcard Type')
plt.xlabel('Count')
plt.title('Count of Each Railcard Type')
plt.tight_layout()
plt.show()
```

```python
railcard_sales = Railway.groupby('Railcard')['Price'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(y=railcard_sales.index, x=railcard_sales.values, orient='h')
plt.ylabel('Railcard Type')
plt.xlabel('Sales')
plt.title('Sales of Each Railcard Type')
plt.tight_layout()
plt.show()
```

## Q2/ Which Ticket Type contributes the most to sales?

### Code:

```python
Ticket_Type_Sales = Railway.groupby('Ticket Type')['Price'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=Ticket_Type_Sales.index, y=Ticket_Type_Sales.values)
plt.xlabel('Ticket Type')
plt.ylabel('Sales')
plt.title('Ticket Type Sales')
plt.show()
```

## Q7/ How much is the Refund Amount?

### Code:

```python
def calculate_refund_amount(df):

    refund_requests = df[df['Refund Request'] == 'Yes']
    total_refund_amount = refund_requests['Price'].sum()
    return total_refund_amount

refund_amount = calculate_refund_amount(Railway)
print(f"The total refund amount is: {refund_amount}")
```

### Answer:
The Refund Amount = 38702

## Q8/ What is the percentage of refund amount to total sales

### Code:

```python
total_sales = Railway['Price'].sum()
refund_percentage = (refund_amount / total_sales) * 100
print(f"The percentage of refund amount to total sales is: {refund_percentage:.2f}%")
```

### Answer:
Refund represents 5.22% of total sales

## Q9/ When are the rush hours?

### Code:

```python
Railway['Departure Time'] = pd.to_datetime(Railway['Departure Time'])
Railway['Departure Hour'] = Railway['Departure Time'].dt.hour
journey_counts = Railway['Departure Hour'].value_counts().sort_index()

plt.figure(figsize=(10, 8))
ax = sns.lineplot(x=journey_counts.index, y=journey_counts.values)
ax.set_xticks(journey_counts.index)

plt.xlabel('Hour of Day')
plt.ylabel('Number of Journeys')
plt.title('Journey Frequency by Hour of Day')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Answer:
6 and 18 are the most rush hours

## Q10/ What are the top 10 Departure Stations and Arrival Destinations?

### Code:

```python
top_10_departure = Railway['Departure Station'].value_counts().nlargest(10)
top_10_arrival = Railway['Arrival Destination'].value_counts().nlargest(10)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.barplot(x=top_10_departure.values, y=top_10_departure.index, orient='h')
plt.title('Top 10 Departure Stations')
plt.xlabel('Number of Departures')
plt.subplot(1, 2, 2)
sns.barplot(x=top_10_arrival.values, y=top_10_arrival.index, orient='h')
plt.title('Top 10 Arrival Destinations')
plt.xlabel('Number of Arrivals')
plt.tight_layout()
plt.show()
```

## Q11/ What are the most 10 used routes?

### Code:

```python
route_counts = Railway.groupby(['Departure Station', 'Arrival Destination']).size().reset_index(name='Count')
top_5_routes = route_counts.sort_values('Count', ascending=False).head(10)
top_5_routes
```

```python
station_connections = Railway.groupby(['Departure Station', 'Arrival Destination']).size().reset_index(name='count')

top_10_connections = station_connections.nlargest(10, 'count')

import plotly.graph_objects as go

fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = list(set(top_10_connections['Departure Station']).union(set(top_10_connections['Arrival Destination']))),
      color = "blue"
    ),
    link = dict(
      source = [list(set(top_10_connections['Departure Station']).union(set(top_10_connections['Arrival Destination']))).index(s) for s in top_10_connections['Departure Station']],
      target = [list(set(top_10_connections['Departure Station']).union(set(top_10_connections['Arrival Destination']))).index(s) for s in top_10_connections['Arrival Destination']],
      value = top_10_connections['count']
  ))])

fig.update_layout(title_text="Top 10 Station Connections", font_size=10)
fig.show()
```

## Q12/ Which stations experience the highest frequency of delays and cancellations?

### Code:

```python
Journey_Status_by_Station = Railway.groupby('Departure Station')['Journey Status'].value_counts().unstack(level='Journey Status')
Delay_by_Station = Journey_Status_by_Station['Delayed']
Delay_by_Station = Delay_by_Station.dropna()
Delay_by_Station = Delay_by_Station.sort_values(ascending=False)
Delay_by_Station
```

```python
Journey_Status_by_Station = Railway.groupby('Departure Station')['Journey Status'].value_counts().unstack(level='Journey Status')
Cancellation_by_Station = Journey_Status_by_Station['Cancelled']
Cancellation_by_Station = Cancellation_by_Station.dropna()
Cancellation_by_Station = Cancellation_by_Station.sort_values(ascending=False)
Cancellation_by_Station
```

Answer:
Liverpool Lime Street faces the most number of delays events, where London Paddington	faces the most of cancellation
